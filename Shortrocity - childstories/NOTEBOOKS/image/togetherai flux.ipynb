```python
from together import Together

client = Together()
arr = []
for i in range(1):
    response = client.images.generate(
        prompt="""Arthur the Ant (tiny six inches long, glossy black exoskeleton with shiny red spots, wears a miniature green cap, always has a cheeky grin, moves swiftly with playful hops) scurries through the bustling anthill. His constant companions: Bella the Beetle (bright blue shell, small wings), and the Worker Bees (tiny yellow stripes with fuzzy bodies) assist him in daily tasks. Arthur cracks a joke, making his friends laugh as they carry food. The underground tunnels are vibrant with organized chaos, illuminated by the soft glow of bioluminescent fungi. Light filters through small openings, creating a lively and cheerful environment.""",
    model="black-forest-labs/FLUX.1-schnell-Free",
    width=1024,
    height=768,
    steps=4,
    n=1,
    response_format="b64_json"
)
    arr.append(response.data[0].b64_json)
from IPython.display import display, Image
import base64
for b64 in arr:
    display(Image(data=base64.b64decode(b64)))

```
